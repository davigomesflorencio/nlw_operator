# Implementação da Arquitetura LeNet-5 em PyTorch

Este notebook contém apenas a declaração do modelo LeNet-5, conforme solicitado.

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class LeNet5(nn.Module):
    def __init__(self):
        super(LeNet5, self).__init__()
        
        # Camada 1: Convolucional. Entrada: 1 canal (Grayscale). Saída: 6 canais. Kernel: 5x5.
        # Se a entrada for 28x28, usamos padding=2 para simular o 32x32 do paper original.
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=6, kernel_size=5, stride=1, padding=2)
        
        # Camada 2: Pooling (Max). Kernel: 2x2. Stride: 2.
        self.pool1 = nn.MaxPool2d(kernel_size=2, stride=2)
        
        # Camada 3: Convolucional. Entrada: 6 canais. Saída: 16 canais. Kernel: 5x5.
        self.conv2 = nn.Conv2d(in_channels=6, out_channels=16, kernel_size=5, stride=1)
        
        # Camada 4: Pooling (Max). Kernel: 2x2. Stride: 2.
        self.pool2 = nn.MaxPool2d(kernel_size=2, stride=2)
        
        # Camada 5: Convolucional (atua como uma camada densa se a entrada for 5x5).
        # Entrada: 16 canais. Saída: 120 canais. Kernel: 5x5.
        self.conv3 = nn.Conv2d(in_channels=16, out_channels=120, kernel_size=5, stride=1)
        
        # Camada 6: Totalmente Conectada (FC).
        self.fc1 = nn.Linear(in_features=120, out_features=84)
        
        # Camada 7: Saída (Output). 10 classes.
        self.fc2 = nn.Linear(in_features=84, out_features=10)

    def forward(self, x):
        # C1 -> ReLU -> S2
        x = F.relu(self.conv1(x))
        x = self.pool1(x)
        
        # C3 -> ReLU -> S4
        x = F.relu(self.conv2(x))
        x = self.pool2(x)
        
        # C5 -> ReLU
        x = F.relu(self.conv3(x))
        
        # Flatten (Achatar para entrar na FC)
        x = torch.flatten(x, 1)
        
        # F6 -> ReLU
        x = F.relu(self.fc1(x))
        
        # Output
        logits = self.fc2(x)
        
        return logits

# Instanciando o modelo para conferência
model = LeNet5()
print(model)
```
